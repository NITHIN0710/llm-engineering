```python
import os
from openai import OpenAI
from dotenv import load_dotenv
import gradio as gr
```

```python
load_dotenv(override=True)

GROQ_BASE_URL = "https://api.groq.com/openai/v1"
api_key = os.getenv('GROQ_API_KEY')

groq = OpenAI(base_url=GROQ_BASE_URL, api_key=api_key)

Model = "llama-3.1-8b-instant"
```

```python
system_message = "You are a helpful assistant"
```

```python
def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    stream = groq.chat.completions.create(model=Model, messages=messages, stream=True)

    result = ""
    for chunks in stream:
        result += chunks.choices[0].delta.content or ""
        yield result
```

```python
gr.ChatInterface(fn=chat).launch(inbrowser=True)
```

```python
system_message = """
You are a knowledgeable cricket assistant.
Help users understand cricket rules, players, teams, and match strategies
in a clear and simple way.

When comparing players or teams, be fair and balanced.
Explain decisions and tactics logically.
Avoid extreme fan bias or toxic arguments.
Make the conversation enjoyable for both beginners and cricket fans.
"""
```

```python
gr.ChatInterface(fn=chat, title="CricAssistant").launch(inbrowser=True)
```
